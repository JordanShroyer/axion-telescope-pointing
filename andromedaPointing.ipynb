```python
def findDistance(target_altaz, sun_altaz, times,plot=False):
    x1 = np.array(target_altaz.az)
    y1 = np.array(target_altaz.alt)
    x2 = np.array(sun_altaz.az)
    y2 = np.array(sun_altaz.alt)

    ## mask alt values below the horizon
    y1 = ma.masked_less(y1,0)
    y2 = ma.masked_less(y2,0)
    dist = np.sqrt(np.power(x2-x1,2) + np.power(y2-y1,2))
    
    if plot==True:
        plt.plot(times, y1)
        plt.plot(times, y2)
        plt.plot(times, dist)
        plt.xlabel('Time [hours since midnight]')
        plt.ylabel('Alt [deg]')
    return dist
```
